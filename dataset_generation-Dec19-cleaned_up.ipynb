```python
import wp_src_code as wp

import cv2, math, os, random
import numpy as np

from joblib import parallel_backend, Parallel, delayed
```

```python
def wallpaper_group_generator(source, output, image_size):
    
    start_number = 0
    imagelist = os.listdir(source) 
    
    for i in range(start_number,len(imagelist)):
        
        image_name = imagelist[i][:-4] #name folder in output directory with image name
        folder_saved_to = output + image_name #folder that image results saved to 

        if not os.path.exists(folder_saved_to): 
            os.makedirs(folder_saved_to) # mkdir for folder_saved_to

        angle = random.randint(20,85)
        unit_h = random.randint(30,80)
        lower_bound = int(unit_h/math.tan(math.radians(angle)))
        if lower_bound > 65:
            unit_w = random.randint(lower_bound+15, lower_bound+75)
        else: 
            unit_w = random.randint(lower_bound+15,85)

        ori = cv2.imread(source+imagelist[i]) 
        image = np.copy(ori) # load image used to crop and transform
        
        width = image_size[0]
        height = image_size[1]

        try: 
            p1_1 = wp.p1_rectangular(image,unit_w,unit_h,width,height)
            p1_2 = wp.p1_square(image,unit_w,width,height)
            p1_3 = wp.p1_oblique(image,unit_w,unit_h,angle,width,height) 
            p1_4 = wp.p1_hexagonal(image,unit_w,width,height)
            p1_5 = wp.p1_rhombic(image,unit_w,unit_h,width,height) 

            cv2.imwrite(folder_saved_to+ '/' +'p1_1.png',p1_1)
            cv2.imwrite(folder_saved_to+ '/' +'p1_2.png',p1_2)
            cv2.imwrite(folder_saved_to+ '/' +'p1_3.png',p1_3)
            cv2.imwrite(folder_saved_to+ '/' +'p1_4.png',p1_4)
            cv2.imwrite(folder_saved_to+ '/' +'p1_5.png',p1_5)
        except:
            pass

        try: 
            p2_1 = wp.p2_rectangular(image,unit_w,unit_h,width,height)
            p2_2 = wp.p2_square(image,unit_w,width,height)
            p2_3 = wp.p2_oblique(image,unit_w,unit_h,angle,width,height) 
            p2_4 = wp.p2_hexagonal(image,unit_w,width,height)
            p2_5 = wp.p2_rhombic(image,unit_w,unit_h,width,height)

            cv2.imwrite(folder_saved_to+ '/' +'p2_1.png',p2_1)
            cv2.imwrite(folder_saved_to+ '/' +'p2_2.png',p2_2)
            cv2.imwrite(folder_saved_to+ '/' +'p2_3.png',p2_3)
            cv2.imwrite(folder_saved_to+ '/' +'p2_4.png',p2_4)
            cv2.imwrite(folder_saved_to+ '/' +'p2_5.png',p2_5)
        except:
            pass


        try: 
            pm_1 = wp.pm_horizontal(image,unit_w,unit_h,width,height)
            pm_2 = wp.pm_vertical(image,unit_w,unit_h,width,height)
            cv2.imwrite(folder_saved_to+ '/' +'pm_1.png',pm_1)
            cv2.imwrite(folder_saved_to+ '/' +'pm_2.png',pm_2)
        except:
            pass


        try: 
            pg_1 = wp.pg_horizontal(image,unit_w,unit_h,width,height)
            pg_2 = wp.pg_vertical(image,unit_w,unit_h,width,height)
            cv2.imwrite(folder_saved_to+ '/' +'pg_1.png',pg_1)
            cv2.imwrite(folder_saved_to+ '/' +'pg_2.png',pg_2)
        except:
            pass

        try: 
            pmm_1 = wp.pmm(image,unit_w,unit_h,width,height)
            cv2.imwrite(folder_saved_to+ '/' +'pmm_1.png',pmm_1)
        except:
            pass

        try: 
            pmg_1 = wp.pmg_horizontal(image,unit_w,unit_h,width,height)
            pmg_2 = wp.pmg_vertical(image,unit_w,unit_h,width,height)
            cv2.imwrite(folder_saved_to+ '/' +'pmg_1.png',pmg_1)
            cv2.imwrite(folder_saved_to+ '/' +'pmg_2.png',pmg_2)
        except:
            pass

        try: 
            cm_1 = wp.cm_horizontal(image,unit_w,unit_h,width,height) 
            cm_2 = wp.cm_vertical(image,unit_w,unit_h,width,height)
            cv2.imwrite(folder_saved_to+ '/' +'cm_1.png',cm_1)
            cv2.imwrite(folder_saved_to+ '/' +'cm_2.png',cm_2)
        except:
            pass

        try: 
            pgg_1 = wp.pgg(image,unit_w,unit_h,width,height)
            cv2.imwrite(folder_saved_to+ '/' +'pgg_1.png',pgg_1) 
        except:
            pass

        try: 
            cmm_1 = wp.cmm(image,unit_w,unit_h,width,height)
            cv2.imwrite(folder_saved_to+ '/' +'cmm_1.png',cmm_1)  
        except:
            pass


        try: 
            p4_1 = wp.p4(image,unit_w,width,height)
            cv2.imwrite(folder_saved_to+ '/' +'p4_1.png',p4_1)  
        except:
            pass


        try: 
            p4m_1 = wp.p4m(image,unit_w,width,height)
            cv2.imwrite(folder_saved_to+ '/' +'p4m_1.png',p4m_1)
        except:
            pass

        try: 
            p4g_1 = wp.p4g(image,unit_w,width,height)
            cv2.imwrite(folder_saved_to+ '/' +'p4g_1.png',p4g_1)
        except:
            pass

        try: 
            p3_1 = wp.p3(image,unit_w,width,height)
            cv2.imwrite(folder_saved_to+ '/' +'p3_1.png',p3_1)
        except:
            pass


        try: 
            p3m1_1 = wp.p3m1(image,unit_w,width,height)  
            cv2.imwrite(folder_saved_to+ '/' +'p3m1_1.png',p3m1_1)
        except:
            pass

        try: 
            p31m_1 = wp.p31m(image,unit_w,width,height)
            cv2.imwrite(folder_saved_to+ '/' +'p31m_1.png',p31m_1)
        except:
            pass

        try: 
            p6_1 = wp.p6(image,unit_w,width,height)
            cv2.imwrite(folder_saved_to+ '/' +'p6_1.png',p6_1)
        except:
            pass

        try: 
            p6m_1 = wp.p6m(image,unit_w,width,height) 
            cv2.imwrite(folder_saved_to+ '/' +'p6m_1.png',p6m_1)
        except:
            pass

    return ()
```

```python
# set how many times you want to run throuth raw images 
#(each time will generate at leaset 4676 images for every categories)
times = 1

# set number of cpu_workers:
parallel_num = 1

raw_images_directory = '/home/yig319/Documents/temp_image_results/img_from/'
result_directory = '/home/yig319/Documents/temp_image_results/img_to/'

# mkdir in result_directory
folder_list = [None]*times
for i in range(times): folder_list[i] = 'folder_' + str(i+1) 
    
for folder in folder_list:
    if not os.path.exists(os.path.join(result_directory, folder)):
        os.mkdir(os.path.join(result_directory, folder))
        
# create output_directory by join result_directory with folder name 
output_directory=[None]*times
for i in range(times): output_directory[i] = result_directory + folder_list[i] +'/'
```

```python
image_size = (256, 256)
```

```python
with parallel_backend('multiprocessing', n_jobs=parallel_num):
    Parallel()(delayed(wallpaper_group_generator)(raw_images_directory, output_directory[i], image_size) for i in range(parallel_num))
```
