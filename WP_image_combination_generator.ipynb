```python
for i in range (1,10001):
    
    index_1, index_2, index_3, index_4 = np.random.randint(0, 17, 4) # choose random symmetry image
    
    # define directory for every group
    dir_1 = root + group[index_1] + '/'
    dir_2 = root + group[index_2] + '/'
    dir_3 = root + group[index_3] + '/'
    dir_4 = root + group[index_4] + '/'
    
    # random pick one image from the group
    name_1 = random.choice(os.listdir(dir_1))
    name_2 = random.choice(os.listdir(dir_2))
    name_3 = random.choice(os.listdir(dir_3))
    name_4 = random.choice(os.listdir(dir_4))

    # define the name for saving combination image
    name_list = [str(i), group[index_1], group[index_2], group[index_3], group[index_4]]
    name = '_'.join(name_list) + '.png'
    
    # read image
    img1 = np.array(cv2.imread(dir_1 + name_1)) 
    img2 = np.array(cv2.imread(dir_2 + name_2)) 
    img3 = np.array(cv2.imread(dir_3 + name_3)) 
    img4 = np.array(cv2.imread(dir_4 + name_4)) 
    
    # combine images 
    upper = np.concatenate((img1,img2), 1) 
    down = np.concatenate((img3,img4), 1) 
    combination = np.concatenate((upper,down)) 
    
    # save combined image to dir_save
    cv2.imwrite(dir_save + '/' + name, combination)
```
